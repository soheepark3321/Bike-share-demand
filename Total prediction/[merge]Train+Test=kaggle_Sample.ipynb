# Train + Test = kaggle_Sample

Train.csv 파일과 Test.csv 파일을 합쳐<br>
완전한 2011 - 2012년도의 data 파일인 Sample 파일 생성

### Read Dataset (Test, Score)

1.test.csv<br>
2.(Kaggle)Score_0.24493_submission.csv

test data와 score 중 count data 병합

*(Kaggle)Score_0.24493_submission.csv은 windspeed feature의 0값을 <br>
randomforest 모델로 값을 예측하여 대체한 버전이다.<br>

```python
# test.csv 불러오기

import pandas as pd

test = pd.read_csv("test.csv", parse_dates = ["datetime"])
test.head()
```

```python
# (Kaggle)Score_0.24493_submission.csv 불러오기

score = pd.read_csv("(Kaggle)Score_0.24493_submission.csv", parse_dates = ["datetime"])
score.head()
```

### Drop datetime column in score.csv

* axis = 1 : col
* axis = 0 or x : row

```python
score_only_count = score.drop(['datetime'], axis = 1)
```

```python
score_only_count.head()
```

# Test.csv + score.csv = complete_test

test.csv과 count만 존재하는 score.csv파일 병합

```python
complete_test = pd.concat([test, score_only_count], axis = 1)
```

```python
complete_test.head()
```

# Train.csv + complete_test = bike_data_2011_2012

train.csv와 complete_test 파일 병합<br>
→ bike_data_2011_2012 : 2011 - 2012 datetime ~ count까지 완전한 데이터 집합

### Read dataset (train)

```python
train = pd.read_csv("train.csv", parse_dates = ["datetime"])
train.head()
```

### Drop casual, registered column in train.csv

```python
train = train.drop(['casual', 'registered'], axis = 1)
train.head()
```

### concat train, complete_test = bike_data_2011_2012

```python
bike_data_2011_2012 = pd.concat([train, complete_test], ignore_index = True)
```

```python
bike_data_2011_2012.head()
```

```python
bike_data_2011_2012.head()
```

### bike_data_2011_2012.csv file save

index = None
row data 맨 좌측 Number index 생성 X

```python
bike_data_2011_2012.to_csv("kaggle_bike_data_2011_2012.csv", index = None)
```

# Sort dataset in bike_data_2011_2012.csv

### Read dataset (bike_data_2011_2012.csv)

```python
bike_data_2011_2012 = pd.read_csv("kaggle_bike_data_2011_2012.csv", parse_dates = ["datetime"])
bike_data_2011_2012.head()
```

### Split datetime to date, time column

date : Year - month - day<br>
time : Hour - Minute - second

```python
bike_data_2011_2012["date"] = [d.date() for d in bike_data_2011_2012['datetime']]
bike_data_2011_2012["time"] = [d.time() for d in bike_data_2011_2012['datetime']]
```

```python
bike_data_2011_2012.head()
```

### Change column order

맨 우측에 있는 date, time col을 좌측 datetime 다음으로 순서 변경

```python
# copy & paste 위해 col list 출력

cols = list(bike_data_2011_2012.columns.values)
print(cols)
```

```python
# col 순서 변경

bike_data_2011_2012 = bike_data_2011_2012[['datetime', 'date', 'time', 'season', 'holiday', 
                                           'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'count']]
```

```python
bike_data_2011_2012.head()
```

### Drop 'datetime' column in dike_data_2011_2012

```python
bike_data_2011_2012 = bike_data_2011_2012.drop(['datetime'], axis = 1)
bike_data_2011_2012.head()
```

### Sort by date first and then by time last

<정렬 순서><br>
1. 날짜<br>
2. 시간

*오름차순

```python
bike_data_2011_2012_sort = bike_data_2011_2012.sort_values(by = ["date", "time"], ascending = [True, True])
```

```python
bike_data_2011_2012_sort.head()
```

### Save as 'sample.csv'

```python
bike_data_2011_2012_sort.to_csv("kaggle_sample.csv", index = None)
```
